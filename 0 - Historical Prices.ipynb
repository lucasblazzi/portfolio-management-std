```python
import pandas as pd
import yfinance as yf
from yahooquery import Ticker



def historical_yfinance(symbols, start_date, end_date):
    _symbols = " ".join(symbols)
    df = yf.download(_symbols, start=start_date, end=end_date)
    df = df[["Close"]].droplevel(axis=1, level=0)
    return df
    
def historical_yquery(symbols, start_date, end_date):
    _symbols = " ".join(symbols)
    tickers = Ticker(_symbols, asynchronous=True)
    df = tickers.history(start=start_date, end=end_date)
    return df

y_fin = historical_yfinance(["BPAC11.SA", "^BVSP"], "2019-01-01", "2020-01-01")
y_query = historical_yquery(["BPAC11.SA", "^BVSP"], "2019-01-01", "2020-01-01")
```
